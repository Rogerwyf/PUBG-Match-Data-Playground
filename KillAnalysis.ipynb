```python
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
stat0 = pd.read_csv('deaths/kill_match_stats_final_0.csv', sep = ',')
```

```python
list(stat0)
```

```python
stat0.shape
```

```python
time5 = stat0[(stat0['time'] >= 0) & (stat0['time'] <= 300)]
time10 = stat0[(stat0['time'] >= 300) & (stat0['time'] <= 600)]
time15 = stat0[(stat0['time'] >= 600) & (stat0['time'] <= 900)]
time20 = stat0[(stat0['time'] >= 900) & (stat0['time'] <= 1200)]
time25 = stat0[(stat0['time'] >= 1200) & (stat0['time'] <= 1500)]
```

```python
ARlist = ['SCAR-L','M416','AKM','M16A4']
SRlist = ['Kar98k','SKS','Mini 14','Mk14']
```

```python
ARsum5 = list()
for AR in ARlist:
    Count = time5['killed_by'][time5['killed_by'] == AR].count()
    ARsum5.append(Count)
print(ARsum5)
ARsum10 = list()
for AR in ARlist:
    Count = time10['killed_by'][time10['killed_by'] == AR].count()
    ARsum10.append(Count)
print(ARsum10)
ARsum15 = list()
for AR in ARlist:
    Count = time15['killed_by'][time15['killed_by'] == AR].count()
    ARsum15.append(Count)
print(ARsum15)
ARsum20 = list()
for AR in ARlist:
    Count = time20['killed_by'][time20['killed_by'] == AR].count()
    ARsum20.append(Count)
print(ARsum20)
ARsum25 = list()
for AR in ARlist:
    Count = time25['killed_by'][time25['killed_by'] == AR].count()
    ARsum25.append(Count)
print(ARsum25)
```

```python
labels = ARlist
sizes = ARsum5
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
